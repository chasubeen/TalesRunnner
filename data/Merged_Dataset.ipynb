## Validation 데이터셋 병합 및 확인

- `captions_val.csv` (**file1**) → `id`, `img_path`, `caption` 포함
- `annotations_val.csv` (**file2**) → 추가 속성(`srcText`, `name`, `i_action`, `classification`, `character`, `setting`, `action`, `feeling`, `causalRelationship`, `outcomeResolution`, `prediction`) 포함
- 두 파일을 `id` 기준으로 병합하여 `dataset_val.csv` (**output_path**) 생성

```python
import pandas as pd

# CSV 파일 경로 설정
file1_path = "/content/drive/My Drive/7th-project/data/captions_val.csv"
file2_path = "/content/drive/My Drive/7th-project/data/annotations_val.csv"
output_path = "/content/drive/My Drive/7th-project/data/dataset_val.csv"  # 저장할 파일 경로

# CSV 파일 불러오기
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

# id 컬럼을 기준으로 병합 (첫 번째 파일의 컬럼이 먼저 오도록)
merged_df = df1.merge(df2, on="id", how="inner")  # inner join (공통 id만 유지)

# 병합된 데이터를 CSV 파일로 저장 (인덱스 없이 저장)
merged_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"병합된 파일이 '{output_path}' 경로에 저장되었습니다.")
```

## Training 데이터셋 병합 및 확인

1) Train CSV 파일 병합

- 실행 시간을 줄이기 위해 두 개의 계정에서 나눠 실행 → `captions_train.csv`, `captions_train2.csv` 생성
- 두 파일을 하나의 데이터로 병합 (아래로 연결) → 병합된 파일 = `merged_captions_train.csv`


```python
# CSV 파일 경로
file1_path = '/content/drive/MyDrive/7th-project/data/captions_train.csv'
file2_path = '/content/drive/MyDrive/7th-project/data/captions_train2.csv'

output_path = "/content/drive/MyDrive/7th-project/data/merged_captions_train.csv"  # 병합된 파일 저장 경로

# 파일 읽기
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path, header=0) # 헤더 제외하고 데이터만 가져오기

# 두 파일 병합
merged_df = pd.concat([df1, df2], ignore_index=True)

# 결과 저장
merged_df.to_csv(output_path, index=False)

# 병합된 데이터 확인
merged_df.head(10)
```

```python
# 파일 개수 확인
merged_df.shape
```

```python
### 중복 데이터 확인 및 삭제

def find_and_remove_duplicate_ids(csv_file_path):
    if not csv_file_path or not os.path.exists(csv_file_path):
        print(f"[ERROR] 파일을 찾을 수 없습니다: {csv_file_path}")
        return

    # CSV 파일 읽기
    df = pd.read_csv(csv_file_path)

    # 'id' 열이 존재하는지 확인
    if 'id' not in df.columns:
        print("[ERROR] CSV 파일에 'id' 열이 존재하지 않습니다.")
        return

    # 중복된 id 찾기 (전체 개수 확인용)
    duplicate_ids = df[df.duplicated(subset=['id'], keep=False)]['id'].tolist()

    if duplicate_ids:
        print(f"[INFO] 중복된 id 개수: {len(set(duplicate_ids))}")
        print(f"[INFO] 중복된 id 리스트 (일부): {list(set(duplicate_ids))[:5]} ...")

        # 중복된 id 제거 (첫 번째 항목만 남기고 뒤의 데이터 삭제)
        df = df.drop_duplicates(subset=['id'], keep='first')

        # 변경된 데이터 저장 (덮어쓰기)
        df.to_csv(csv_file_path, index=False, encoding='utf-8')

        print(f"[SUCCESS] 중복 제거 완료! {csv_file_path} 업데이트됨.")
    else:
        print("[INFO] 중복된 id가 없습니다.")


find_and_remove_duplicate_ids(output_path)
```
